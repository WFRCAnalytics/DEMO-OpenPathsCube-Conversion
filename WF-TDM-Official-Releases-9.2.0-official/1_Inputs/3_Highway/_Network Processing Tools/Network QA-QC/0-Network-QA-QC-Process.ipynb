# Network QA-QC Process

The steps herein detailed is the QA-QC process to be undertaken upon the completion of highway network changes prior to general distribution to potential model users.

## QA-QC Process

- Run `1-Network-QA-QC-Checks.ipynb` in this folder
    - Check for inconsistent `LN` values changes between scenario-years (goes from >0 to 0 and back, etc)
    - Check where `LN` fields are zero in all years (link not being used at all, maybe have exception list)
    - Check where nodes are not connected to anything (no corresponding record in link dataset)
    - Check where `FT` fields are zero but corresponding `LN` fields not zero
    - Check where `LN` fields are zero but corresponding `FT` fields not zero
    - Check if link is centroid connecter and that `FT`=1 and `LN`=7
    - Check if transit coded on links without SEGID (other than freeways or cross-streets between freeway ramp terminals)
- Run the segment checks in `1_Inputs\6_Segment\_Segment Processing Tools\1-Segment-Checks.ipynb`
    - Maybe add some more checks to that script:
        - Check if SEGID in multiple counties
        - Check if segment is within XX distance from TDM links
- Check that transit compiles for each scenario: Run the `_chk Transit Compile on Net` scripts for each model scenario
- Test run each model
    - Check if model gets to distribution:
    - Check if there are any unconnected TAZs in LOG file


As part of coding the highway network with SEGIDs for transit:
  - All links (excluding freeways) over which transit travels shall have a SEGID
  - Centroid connectors that connect rail stops with roadway network (over which there is a bus route) shall have a SEGID
