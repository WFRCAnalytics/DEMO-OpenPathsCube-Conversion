# Definitions

**LRS Segments:** segments with a numeric prefix in SEGID. Prefix is < 1000 for state roads and >1000 for federal-aid routes (non-state roads that recieve federal funding)

**Planning Segments:** segments with a non-numeric prefix in SEGID. Valid prefixes for WFRC/MAG area are `MAG`, `UTA`, `WFRC`.

**UTA Segments:** a subset of *Planning Segments*, these segments have IDs that follow a pattern (see **FILE????**)

# Data Files

**AADTs**: UDOT's AADT layer service https://maps.udot.utah.gov/central/rest/services/AADT/AADT_Unrounded/MapServer/0

**Counties** UGRC's Utah County Boundaries layer service: https://services1.arcgis.com/99lidPhWCzftIe9K/arcgis/rest/services/UtahCountyBoundaries/FeatureServer/0

**Factors** latest day-of-week, month, and season factors data from UDOT-Segment-Factors repo: https://github.com/WFRCAnalytics/UDOT-Segment-Factors

**Forecasts** latest AADT forecasts from Traffic-Volume-Forecasts repo: https://github.com/WFRCAnalytics/Traffic-Volume-Forecasts

**MasterNet** latest version of master network saved as a link shapefile

**Segments:** latest version of segments (with only fields of `SEGID`, `BMP`, `EMP`, `Distance`). Named according to [UDOT Standard Naming for Common TDM Files](https://docs.google.com/document/d/1MaglBpVndBrAZd4_COkfUdBefNMjua-CjEapXwXf3n0/edit). Wastach Front Example: `WF` + [version] + `_Segments` + `.shp`, (eg. `WFv910_Segments.shp`)



# How to Prep Segments for TDM

**Step 0:** Prepare Segments in Shapefile and Code SegIds into TDM

Use `_arcgispro` project to help match TDM and segment shapefile.

Good idea to keep track of changes when editing segment shapefile and master network. (Not sure if a formal location/format should be adopted)

For *LRS Segments* :

 - Must be drawn within 2 meters of AADTs segments. (for spatial join of centroid to AADT layer)
 - BMP, EMP, Distance are also rough estimates, despite precision seen.
 - BMP is used as suffix in SEGID, but rounded to the tenth of a mile.

For *Planning Segments*

 - Numbered sequentially, always added new numbers at the end
 - Avoid reusing a SEGID that has been retired, due to potential conflic with past datasets

UTA Segment numbering... (add google sheet to repo)

**Step 1:** Check TDM and Segments Harmonization (`1-SegmentChecks.ipynb`)

  1. Unique SEGID in Segment Shapefile
  2. Distance in Shapefile = Calculated Distance
  3. All *LRS Segments*: `Distance = EMP-BMP` (to nearest tenth of a mile)
  3. All TDM SEGIDs in Segments Shapefile (except exclusions coded into `params/segid-exemptions-notInSegments.csv`)
  4. All Segments Shapefile SEGIDs in TDM (except exclusions coded into `params/segid-exemptions-notInTdm.csv`)
  5. Check coordinate system and (??)

**Step 2:** Add Geographic Keys and Direction onto Segments (`2-Add-Geo-Keys-Direction.inpynb`):

 - SUBAREAID
 - PLANAREA
 - CO_FIPS
 - DIRECTION

**Step 3:** Add AADTs onto Segments (`2-Add-AADTs.inpynb`)

For accurate joining purposes, the centroids of all *LRS Segments* must be within 2 meters of the UDOT AADT layer.

**Step 4:** Add Factors from Continous Count Station groups (`3-Assign-Factors-to-Segments.ipynb`)

**Step 5:** Add final forecasts from forecast process (`Run 4-Assign-Forecasts-to-Segments.ipynb`)
